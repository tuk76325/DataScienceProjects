```python
import numpy as np
import pandas as pd

# load data with pandas
df = pd.read_csv('cars.csv')
# print(df)

# convert a pandas dataframe to a numpy array
X = df.values[:, 1:-1]
print(type(X))
print(X.shape)
print(X)
```

```python
#Get first row and the second to last row

x1 = X[0, :]
x2 = X[-2, :]

print(x1)
print(x2)

# Addition

y = x1 + x2
print(y)
```

```python
# Get first 5 rows and last 5 rows 

m1 = X[0:5, :]
m2 = X[-5:, :]

print(m1)
print(m2)

# Subtraction
m = m1 - m2
print(m)
```

```python
# this is your code to get two vectors

x1 = X[:, 0]
x2 = X[:, -2]

print(x1)
print(x2)
```

```python
#1. addition and subtraction

x1 = X[:, 0]
x2 = X[:, -2]

y = x1 + x2
z = x1 - x2

#2. element-wise multiplication

y2 = np.multiply(x1, x2)
print(y2)

#3. Inner product

y3 = np.inner(x1, x2)
print(y3)

# 4. three norms

y4 = np.linalg.norm(x1, ord=1)
print(y4)
y5 = np.linalg.norm(x1, ord=2)
print(y5)
y6 = np.linalg.norm(x1, ord=np.inf)
print(y6)

# 5. compute the distance between these two vectors

y7 = np.linalg.norm(x2-x1, ord=2)
print(y7)
```

```python
# this is your code to get two matrices: A, B

A = X[0:2, :]
B = X[3:5, :]

print(A)
print(B)
```

```python
# 1. addition and subtraction 

print(A + B)
print(A - B)

# 2. compute the transpose of the second matrix: C = B^T, and print the shape of C

C = np.transpose(B)
print(str(len(C)) + ', ' + str(len(C[0])))

# 3. compute the product between A and C, and print the result and its shape

dot = np.dot(A, C)
print(dot)
print(str(len(dot)) + ', ' + str(len(dot[0])))

# 4. compute the transpose of the first matrix: D = A^T, and print the shape of D

D = np.transpose(A)
print(str(len(D)) + ', ' + str(len(D[0])))

# 5. compute the product between D and B, and print the result and its shape

dot1 = np.dot(D, B)
print(dot1)
print(str(len(dot1)) + ', ' + str(len(dot1[0])))
```
